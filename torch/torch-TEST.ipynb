# Tensor 张量测试

```python
import torch
```

```python
tensor0d = torch.tensor(1)
print(tensor0d.shape, tensor0d.dtype)
```

```python
tensor1d = torch.tensor([1, 2, 3])
print(tensor1d.shape, tensor1d.dtype)
```

```python
tensor2d = torch.tensor([[1, 2, 3], [4., 5., 6.]])
print(tensor2d.shape, tensor2d.dtype)
```

```python
tensor3d = torch.tensor([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(tensor3d.shape, tensor3d.dtype)

float3d = tensor3d.to(torch.float32)
print(float3d.shape, float3d.dtype)
```

```python
x = tensor3d.stride()
print(x)

# (4, 2, 1)
# 跳到下一块，需要跨4个数；跳到下一行，需要跨2个数；跳到下一列，需要跨1个数。
```

```python
float2d = torch.tensor([[1, 2, 3], [4., 5., 6.]], dtype=torch.float32)
print(float2d.shape, float2d.dtype)
```

```python
float64 = torch.tensor([[1, 2, 3], [4., 5., 6.]], dtype=torch.float64)
print(float64.shape, float64.dtype)
```

# Tensor 重塑

```python
import torch
```

```python
tensor2d = torch.tensor([[1, 2, 3], [7, 8, 9]], dtype=torch.float32)
print(tensor2d.shape, tensor2d.dtype)
print(tensor2d)
```

```python
view2d = tensor2d.view(3, 2)
print(view2d.shape, view2d.dtype)
print(view2d)
```

```python
reshape2d = tensor2d.reshape(3, 2)
print(reshape2d.shape, reshape2d.dtype)
print(reshape2d)
```

## 张量转置

```python
tensor2T = tensor2d.T
print(tensor2T.shape, tensor2T.dtype)
print(tensor2T)
```

## 张量相乘

```python
print(tensor2d)
print(tensor2T)

matmul2d = tensor2d @ tensor2T
print(matmul2d.shape, matmul2d.dtype)
print(matmul2d)

matmul2d_2 = torch.matmul(tensor2d, tensor2T)
print(matmul2d_2.shape, matmul2d_2.dtype)
print(matmul2d_2)
```
