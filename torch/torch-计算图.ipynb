# 计算图

```python
import torch
import torch.nn.functional as F
```

```python
# 定义所有变量
y  = torch.tensor([1.0])   # 真实标签（目标值）
x1 = torch.tensor([1.1])   # 输入
w1 = torch.tensor([2.2])   # 权重
b  = torch.tensor([0.0])   # 偏置

# 前向计算
z = w1 * x1 + b       # 加权和
a = torch.sigmoid(z)  # 激活输出（预测概率）
loss = F.binary_cross_entropy(a, y)  # 二分类交叉熵损失

print('y=', y)
print('x1=', x1)
print('w1=', w1)
print('b=', b)
print('z=', z)
print('a=', a)
print('loss=', loss)
```

# 自动微分

```python
import torch
import torch.nn.functional as F
from torch.autograd import grad
```

```python
# 定义所有变量
y  = torch.tensor([1.0])   # 真实标签（目标值）
x1 = torch.tensor([1.1])   # 输入
w1 = torch.tensor([2.2], requires_grad=True)   # 权重
b  = torch.tensor([0.0], requires_grad=True)   # 偏置

# 前向计算
z = w1 * x1 + b       # 加权和
a = torch.sigmoid(z)  # 激活输出（预测概率）
loss = F.binary_cross_entropy(a, y)  # 二分类交叉熵损失

# 计算梯度
grad_L_w1, grad_L_b = grad(loss, [w1, b], retain_graph=True)
print('grad_L_w1=', grad_L_w1)
print('grad_L_b=', grad_L_b)

# 反向传播
loss.backward()
print('w1.grad=', w1.grad)
print('b.grad=', b.grad)
```
